# Permutaciones

```python
from itertools import permutations
```

## Calculando las permutaciones de tamaño n

```python
S=["H1", "H2","M1","M2"]
perm = permutations(S)
for i in list(perm):
    print (i)
```

## Calculando las permutaciones de tamaño r de un total de n.

```python
S=[1, 2,3,4]
```

```python
# Obtener todas las permutaciones de tamaño 2
S=[1, 2,3,4]
perm = permutations(S, 3)
for i in list(perm):
    print (i)
```

# Combinaciones

```python
from itertools import combinations
```

## Calculando las combinaciones de tamaño r de un total de n.

```python
S=[1,2,3,4,5]
comb = combinations(S, 2)
for i in list(comb):
    print (i)
```

# Combinaciones con Reposición

```python
from itertools import combinations_with_replacement
```

## Calculando las combinaciones con reposición de tamaño r de un total de n.

```python

S=[1,2,3,4]
comb = combinations_with_replacement(S, 2)
for i in list(comb):
    print (i)
```

# Funciones en Python

## Factorial

```python
# Factorial
def fact(n):
  if n == 0:
    return 1
  if n == 1:
    return 1
  else:
    return n * fact(n-1)
```

```python
fact(5)
```

```python
fact(0)
```

```python
fact(8)/fact(6)
```

## Permutación

```python
# Permutación
def Perm(n,k):
  if n <k :
    return "No valido"
  else:
    return fact(n)/fact(n-k)
```

```python
Perm(5,2)
```

```python
Perm(5,5)
```

## Combinación

```python
# Combinación
def Comb(n,k):
  if n <k :
    return "No valido"
  else:
    return fact(n)/(fact(k)*fact(n-k))
```

```python
Comb(6,2)
```

```python
Comb(6,3)
```

```python
Comb(6,2)+Comb(6,3)
```

```python
Comb(7,3)
```

```python
Comb(6,3)*Comb(4,0)/Comb(10,3)
```
